```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the CSV files
adaptive_results = pd.read_csv('adaptive-results.csv')
full_results = pd.read_csv('full-results.csv')

# Extract the relevant data
iterations_adaptive = adaptive_results['iteration']
dt_adaptive = adaptive_results['iteration-dt']

iterations_full = full_results['iteration']
dt_full = full_results['iteration-dt']

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(iterations_adaptive, dt_adaptive, '-g', label='Adaptive Method', lw=4)
plt.plot(iterations_full, dt_full, '-b', label='Full Method', lw=4)

# Label the axes
plt.xlabel('Iteration Count', size=16)
plt.ylabel('Iteration Time (sec)', size=16)
plt.tick_params(axis='both', labelsize=12)
plt.ylim([0, 15])

# Title and legend
# plt.title('Comparison of Iteration Time: Full vs. Adaptive Method')
plt.legend(loc='upper left', prop={'size': 14})
plt.grid()

# Show the plot
plt.savefig('full-vs-adaptive-time-comparison-no-title.pdf')
plt.savefig('full-vs-adaptive-time-comparison-no-title.png')
plt.show()
```

```python
# Stacked bar for the adaptive method
plt.figure(figsize=(12,6))
plt.bar(adaptive_results['iteration'], adaptive_results['refine-dt'], label='Refine', color='skyblue')
plt.bar(adaptive_results['iteration'], adaptive_results['build-dt'], bottom=adaptive_results['refine-dt'], label='Build', color='blue')
plt.bar(adaptive_results['iteration'], adaptive_results['upwards-dt'], bottom=adaptive_results['refine-dt'] + adaptive_results['build-dt'], label='Upwards', color='green')
plt.bar(adaptive_results['iteration'], adaptive_results['solve-dt'], bottom=adaptive_results['refine-dt'] + adaptive_results['build-dt'] + adaptive_results['upwards-dt'], label='Solve', color='red')
plt.xlabel('Iteration', size=16)
plt.ylabel('Time (seconds)', size=16)
plt.tick_params(axis='both', labelsize=12)
plt.legend(loc='lower left', prop={'size': 14})
plt.tight_layout()
plt.savefig('adaptive-stacked-bar-no-title.pdf')
plt.savefig('adaptive-stacked-bar-no-title.png')
plt.show()

# Stacked bar for the full method
plt.figure(figsize=(12,6))
plt.bar(full_results['iteration'], full_results['refine-dt'], label='Refine', color='skyblue')
plt.bar(full_results['iteration'], full_results['build-dt'], bottom=full_results['refine-dt'], label='Build', color='blue')
plt.bar(full_results['iteration'], full_results['upwards-dt'], bottom=full_results['refine-dt'] + full_results['build-dt'], label='Upwards', color='green')
plt.bar(full_results['iteration'], full_results['solve-dt'], bottom=full_results['refine-dt'] + full_results['build-dt'] + full_results['upwards-dt'], label='Solve', color='red')
plt.xlabel('Iteration', size=16)
plt.ylabel('Time (seconds)', size=16)
plt.tick_params(axis='both', labelsize=12)
plt.legend(loc='lower left', prop={'size': 14})
plt.tight_layout()
plt.savefig('full-stacked-bar-no-title.pdf')
plt.savefig('full-stacked-bar-no-title.png')
plt.show()
```
